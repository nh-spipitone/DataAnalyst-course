```python
import pandas as pd
from sqlalchemy import create_engine, text

# Creazione connessione al database PostgreSQL
engine = create_engine("postgresql+psycopg2://postgres:mypassword@localhost:5432/esercizio")
```

```python
# filtra i pokemon di tipo "fire" (solo mono tipo)
fire_pokemon_query = text("SELECT * FROM pokemon WHERE types = 'fire'")
fire_pokemon = pd.read_sql(fire_pokemon_query, engine)
fire_pokemon
```

```python
# filtra i pokemon di tipo "fire" (anche multi tipo)
fire_multi_pokemon_query = text("SELECT * FROM pokemon WHERE types LIKE '%fire%'")
fire_multi_pokemon = pd.read_sql(fire_multi_pokemon_query, engine)
fire_multi_pokemon
```

```python
# filtra i pokemon con hp maggiore di 170
hp_query = text("""
	SELECT * FROM pokemon
	WHERE CAST(
		regexp_replace(stats, '^.*hp=([0-9]+).*$','\\1') AS INTEGER
	) > 170
""")
hp_pokemon = pd.read_sql(hp_query, engine)
hp_pokemon[['name', 'types', 'stats']]
```

```python
# printa la colonna stats per intero per il pokemon chansey
chansey_stats_query = text("""
    SELECT stats FROM pokemon
    WHERE name = 'chansey'
""")
chansey_stats = pd.read_sql(chansey_stats_query, engine)
chansey_stats['stats'].values[0]  # Stampa la colonna stats per Chansey
```

```python
# accedi al database splittando le stats in colonne separate
stats_query = text("""
    SELECT
        *,
        CAST(regexp_replace(stats, '^.*hp=([0-9]+).*$','\\1') AS INTEGER) AS hp,
        CAST(regexp_replace(stats, '^.* attack=([0-9]+).*$','\\1') AS INTEGER) AS attack,
        CAST(regexp_replace(stats, '^.* defense=([0-9]+).*$','\\1') AS INTEGER) AS defense,
        CAST(regexp_replace(stats, '^.*special-attack=([0-9]+).*$','\\1') AS INTEGER) AS sp_attack,
        CAST(regexp_replace(stats, '^.*special-defense=([0-9]+).*$','\\1') AS INTEGER) AS sp_defense,
        CAST(regexp_replace(stats, '^.*speed=([0-9]+).*$','\\1') AS INTEGER) AS speed,
        CAST(regexp_replace(stats, '^.*hp=([0-9]+).*$','\\1') AS INTEGER) +
        CAST(regexp_replace(stats, '^.* attack=([0-9]+).*$','\\1') AS INTEGER) +
        CAST(regexp_replace(stats, '^.* defense=([0-9]+).*$','\\1') AS INTEGER) +
        CAST(regexp_replace(stats, '^.*special-attack=([0-9]+).*$','\\1') AS INTEGER) +
        CAST(regexp_replace(stats, '^.*special-defense=([0-9]+).*$','\\1') AS INTEGER) +
        CAST(regexp_replace(stats, '^.*speed=([0-9]+).*$','\\1') AS INTEGER) AS total_stats
    FROM pokemon
""")
pokemon_stats_df = pd.read_sql(stats_query, engine)
pokemon_stats_df.head(20)
```

```python
# carica il DataFrame nel database
pokemon_stats_df.to_sql('pokemon', engine, if_exists='replace', index=False)
# salva il dataframe in un file sqlite
pokemon_stats_df.to_sql('pokemon', 'sqlite:///pokemon.db', if_exists='replace', index=False)
```

```python
# filtra i Pokémon fire con HP superiore a 70
fire_hp_query = text("""
    SELECT * FROM pokemon WHERE types LIKE '%fire%' AND hp > 70
""")
fire_hp_pokemon = pd.read_sql(fire_hp_query, engine)
fire_hp_pokemon
```

```python
# mostra il pokemon con più alto attacco
highest_attack_query = text("""
    SELECT name, types, hp, attack, defense, sp_attack, sp_defense, speed, total_stats
    FROM pokemon
    ORDER BY attack DESC
    LIMIT 1
""")
highest_attack_pokemon = pd.read_sql(highest_attack_query, engine)
highest_attack_pokemon
```

```python
# mostra le stats del pokemon "moltres"
# accedi al database splittando le stats in colonne separate
stats_query = text("""
    SELECT name, types, hp, attack, defense, sp_attack, sp_defense, speed, total_stats
    FROM pokemon
    WHERE name = 'moltres'
""")
pokemon_stats_df = pd.read_sql(stats_query, engine)
pokemon_stats_df.head()
```

```python
# crea una funzione che accetta un nome di Pokémon e restituisce le sue stats
def get_pokemon_stats(pokemon_name):
    query = text(f"""
        SELECT name, types, hp, attack, defense, sp_attack, sp_defense, speed, total_stats
        FROM pokemon
        WHERE name = :pokemon_name
    """)
    return pd.read_sql(query, engine, params={"pokemon_name": pokemon_name})

# Esempio di utilizzo della funzione
get_pokemon_stats('rayquaza')
```

```python
# crea una top 10 dei Pokémon con le migliori stats totali
top_50_query = text("""
    SELECT name, types, hp, attack, defense, sp_attack, sp_defense, speed, total_stats
    FROM pokemon
    ORDER BY total_stats DESC
    LIMIT 10
""")
top_50_pokemon = pd.read_sql(top_50_query, engine)
top_50_pokemon
```

```python
# crea una top 10 dei Pokémon con le peggiori stats totali
top_50_query = text("""
    SELECT name, types, hp, attack, defense, sp_attack, sp_defense, speed, total_stats
    FROM pokemon
    ORDER BY total_stats
    LIMIT 10
""")
top_50_pokemon = pd.read_sql(top_50_query, engine)
top_50_pokemon
```
