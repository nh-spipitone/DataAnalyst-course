```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/home/giovanni/Programs/Python3/DataAnalyst-course/Esercizi/Giorno 5/temperature.csv")
df
```

```python
#Converte la colonna "Data" in formato datetime.
df['Data'] = pd.to_datetime(df['Data'])
df.head()
```

```python
#Calcola la temperatura media della prima settimana (dal 1 al 7 gennaio).
df['Settimana'] = df['Data'].dt.isocalendar().week
temperatura_media_settimana = df[df['Settimana'] == 1]['Temperatura'].mean()
print(f"Temperatura media della prima settimana: {temperatura_media_settimana:.2f}째C")
```

```python
#Calcola la temperatura media del periodo analizzato.
temperatura_media_periodo = df['Temperatura'].mean()
print(f"Temperatura media del periodo analizzato: {temperatura_media_periodo:.2f}째C")
```

```python
# Visualizza un grafico a linee dellandamento delle temperature giornaliere.
plt.figure(figsize=(12, 6))
plt.plot(df['Data'], df['Temperatura'], marker='o', linestyle='-', color='blue')
plt.title("Andamento delle temperature giornaliere")
plt.xlabel("Giorno")
plt.ylabel("Temperatura (째C)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(df['Temperatura'].min() * 0.8, df['Temperatura'].max() * 1.15)  # Imposta un range per l'asse y con un certo margine superiore e inferiore
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#(Extra) Individua il giorno pi첫 caldo e quello pi첫 freddo.
giorno_piu_caldo = df.loc[df['Temperatura'].idxmax()]
giorno_piu_freddo = df.loc[df['Temperatura'].idxmin()]

print(f"Giorno pi첫 caldo: {giorno_piu_caldo['Data'].date()} con temperatura {giorno_piu_caldo['Temperatura']}째C")
print(f"Giorno pi첫 freddo: {giorno_piu_freddo['Data'].date()} con temperatura {giorno_piu_freddo['Temperatura']}째C")
```
