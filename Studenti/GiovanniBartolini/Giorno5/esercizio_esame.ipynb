```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("/home/giovanni/Programs/Python3/DataAnalyst-course/Esercizi/Giorno 5/studio_esame.csv")
df
```

```python
#Calcola e stampa la media delle ore di studio
media_studio = df["OreStudio"].mean()
print(f"La media delle ore di studio è: {media_studio:.2f} ore")
```

```python
#Calcola e stampa la media dei punteggi d’esame
media_punteggio = df["PunteggioEsame"].mean()
print(f"La media dei punteggi d'esame è: {media_punteggio:.2f}")
```

```python
# Crea uno scatter plot con: asse x = OreStudio, asse y = PunteggioEsame
plt.figure(figsize=(6, 4))
plt.scatter(df["OreStudio"], df["PunteggioEsame"], color='red', label='Ore di Studio vs Punteggio d\'Esame')
plt.title("Ore di Studio vs Punteggio d'Esame")
plt.xlabel("Ore di Studio")
plt.ylabel("Punteggio d'Esame")
plt.grid(True)
# aggiungi il nome di ogni studente sotto al rispettivo punto
for i, row in df.iterrows():
    plt.annotate(row["Studente"], (row["OreStudio"], row["PunteggioEsame"]), 
                 textcoords="offset points", xytext=(0,-10), ha='center', fontsize=8)
# Evidenzia sul grafico lo studente con il punteggio più alto (marcatore speciale o etichetta).
max_punteggio_index = df["PunteggioEsame"].idxmax()
plt.scatter(df.loc[max_punteggio_index, "OreStudio"], df.loc[max_punteggio_index, "PunteggioEsame"], 
            color='blue', s=100, label='Studente con punteggio più alto')
plt.legend()
plt.annotate(f'Massimo: {df.loc[max_punteggio_index, "PunteggioEsame"]}', 
             (df.loc[max_punteggio_index, "OreStudio"], df.loc[max_punteggio_index, "PunteggioEsame"]),
             textcoords="offset points", xytext=(0,10), ha='center', color='blue')
# Aggiungi una linea di regressione 
z = np.polyfit(df["OreStudio"], df["PunteggioEsame"], 1) 
p = np.poly1d(z)
plt.plot(df["OreStudio"], p(df["OreStudio"]), color='green', label='Linea di Regressione', linestyle='-')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# (Extra) Calcola il coefficiente di correlazione tra ore di studio e punteggio finale e mostra il valore.
correlazione = df["OreStudio"].corr(df["PunteggioEsame"])
print(f"Il coefficiente di correlazione tra ore di studio e punteggio d'esame è: {correlazione:.2f}")
```
