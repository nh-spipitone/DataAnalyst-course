```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

conn = sqlite3.connect('pokemon.db')

df = pd.read_csv('pokemon_data.csv')
df.head()
```

```python
def split_stats_columns(df):
    df["hp"] = df["stats"].str.extract(r"hp=(\d+)").astype(float)
    df["attack"] = df["stats"].str.extract(r"attack=(\d+)").astype(float)
    df["defense"] = df["stats"].str.extract(r"defense=(\d+)").astype(float)
    df["sp_attack"] = df["stats"].str.extract(r"special-attack=(\d+)").astype(float)
    df["sp_defense"] = df["stats"].str.extract(r"special-defense=(\d+)").astype(float)
    df["speed"] = df["stats"].str.extract(r"speed=(\d+)").astype(float)
    return df

split_stats_columns(df)
```

```python
df.to_sql('pokemon', conn, if_exists='replace', index=False)

query1 = "SELECT name, types, hp FROM pokemon WHERE types LIKE '%Fire%' AND hp > 70"
pokemon_tipo_fuoco = pd.read_sql_query(query1, conn)
pokemon_tipo_fuoco
```

```python
query2 = "SELECT name, attack FROM pokemon ORDER BY attack DESC LIMIT 1"
pokemon_attacco_max = pd.read_sql_query(query2, conn)
pokemon_attacco_max
```

```python
query3 = "SELECT name, types, hp, attack, defense, sp_attack, sp_defense, speed FROM pokemon WHERE name LIKE 'Moltres'"
pokemon_moltres = pd.read_sql_query(query3, conn)
pokemon_moltres
```
