Importiamo Pandas, creiamo una funzione carica dati che ci legge un csv e poi richiamiamo la funzione

```python
import pandas as pd

def carica_dati(percorso:str) -> pd.DataFrame:
    df=pd.read_csv(percorso)
    return df

carica_dati("vendite.csv")
```

```python
def riepilogo_vendite(df: pd.DataFrame) -> pd.DataFrame:
    totale_quantita = df["quantita"].sum()
    df["totale_ricavi"] = df["quantita"] * df["prezzo_unitario"]
    totale_ricavi = df["totale_ricavi"].sum()
    ordine_medio = totale_ricavi / len(df)
    
    return pd.DataFrame({"totale_quantita": [totale_quantita],
                         "totale_ricavi": [totale_ricavi],
                         "ordine_medio": [ordine_medio]})
    

riepilogo_vendite(carica_dati("vendite.csv"))
   
    
    
```
