Esercizio 1: Analisi passeggeri Titanic + Istogramma dell'età
Obiettivo: Analizzare e visualizzare la distribuzione dell’età dei passeggeri del Titanic.

Istruzioni:

Carica il dataset titanic.csv.

Rimuovi le righe con età mancante (Age).

Calcola:

L’età media

Il numero totale di passeggeri

Crea un istogramma dell’età (Age) con matplotlib.pyplot.hist().

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")
df
```

```python
df = df.dropna(subset=["Age"])
media_eta = df["Age"].mean()
print(f"La media dell'età dei passeggeri è: {media_eta:.2f} anni")

totale_passeggeri = df["Name"].count()
print(f"Il totale dei passeggeri è: {totale_passeggeri}")

plt.figure(figsize=(8, 6))
plt.hist(df["Age"], bins=10, color='lightblue', edgecolor='black')
plt.title("Distribuzione dell'età dei passeggeri")
plt.xlabel("Età")
plt.ylabel("Numero di passeggeri")
plt.grid(True)
plt.tight_layout()
```

```python
passeggeri_per_sesso = df.groupby("Sex")["Survived"].count()
passeggeri_per_sesso

plt.figure(figsize=(8, 5))
passeggeri_per_sesso.plot(kind='bar', color=['pink', 'blue'])
plt.legend(["Femmine", "Maschi"])
plt.title("Numero di passeggeri per sesso")
plt.xlabel("Sesso")
plt.ylabel("Numero Passeggeri")
plt.grid(True)
plt.xticks(rotation=0)
plt.tight_layout()
```
