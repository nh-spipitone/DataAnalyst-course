```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("temperature.csv")
df
```

```python
df["Data"] = pd.to_datetime(df["Data"])
prima_settimana = df["Data"].dt.day <= 7
media_prima_settimana = df[prima_settimana]["Temperatura"].mean()
print(f"Media della prima settimana: {media_prima_settimana:.2f}°C")
```

```python
media_totale = df["Temperatura"].mean()
print(f"Media totale: {media_totale:.2f}°C")
```

```python
plt.figure(figsize=(10, 8))
plt.plot(df["Data"], df["Temperatura"], marker='o')
plt.xlabel("Data", fontsize=14)
plt.ylabel("Temperatura (°C)", fontsize=14)

```
