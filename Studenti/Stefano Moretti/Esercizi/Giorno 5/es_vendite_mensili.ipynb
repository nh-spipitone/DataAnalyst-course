Esercizio 3: Analisi vendite mensili + Line plot
Obiettivo: Visualizzare l’andamento delle vendite nel tempo.

Istruzioni:

Usa un dataset sales.csv con almeno due colonne: Data e Vendite.

Assicurati che la colonna Data sia in formato datetime.

Raggruppa le vendite per mese (usa resample("M") o estrai month).

Crea un grafico a linee dell’andamento vendite.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales.csv")
df["Data"] = pd.to_datetime(df["Data"])
df["Mese"] = df["Data"].resample("M")

plt.figure(figsize=(10, 6))
df.groupby("Mese")["Vendite"].sum().plot(marker='o')
plt.title("Vendite mensili")
plt.xlabel("Mese")
plt.ylabel("Vendite")
plt.grid(True)
plt.tight_layout()
```

```python
totale_vendite = df["Vendite"].sum()
vendite_mensili = df.groupby("Mese")["Vendite"].sum()

plt.figure(figsize=(10, 6))
vendite_mensili.plot(kind="pie", autopct='%1.1f%%', startangle=90)
plt.title("Distribuzione delle per mese")
plt.ylabel("")
plt.tight_layout()
```
