```python
# Importa le librerie necessarie
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Legge il dataset da un file CSV in un DataFrame pandas
df_dati = pd.read_csv('linear_regression_dataset.csv')

# Estrae la colonna "Feature" come variabile indipendente (X) e "Target" come variabile dipendente (y)
x = df_dati[["Feature"]].values  # X deve essere 2D per scikit-learn
y = df_dati["Target"].values     # y può essere 1D

# Divide il dataset in dati di training e di test (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# Crea e addestra il modello di regressione lineare sui dati di training
model = LinearRegression()
model.fit(x_train, y_train)

# Stampa il coefficiente angolare (pendenza), l'intercetta e il punteggio R^2 sui dati di test
print(f"Coefficente, pendenza: {model.coef_[0]:.3f}")
print(f"Intercetta, ordinata all'origine: {model.intercept_:.3f}")
print(f"R^2: {model.score(x_test, y_test):.3f}")

# Visualizza i dati e la retta di regressione
plt.scatter(x, y, label='Dati', color='blue')  # Dati originali
x_line = x.copy()
x_line.sort(axis=0)                            # Ordina X per una linea continua
y_line = model.predict(x_line)                 # Calcola la retta di regressione
plt.plot(x_line, y_line, color='red', label='Regr. Lineare')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Regr. Lineare')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv('dataset_regressione.csv')
x = df[["OreStudio"]].values  # X deve essere 2D per scikit-learn
y = df["VotoEsame"].values     # y può essere 1D

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

model = LinearRegression()
model.fit(x_train, y_train)

print(f"Coefficiente, pendenza: {model.coef_[0]:.3f}")
print(f"Intercetta, ordinata all'origine: {model.intercept_:.3f}")
print(f"R^2: {model.score(x_test, y_test):.3f}")
print(f"Voto previsto per 5 ore di studio: {model.predict([[5]])[0]:.2f}")

plt.scatter(x, y, label='Dati')  # Dati originali
x_line = x.copy()
x_line.sort(axis=0)                            # Ordina X per una linea continua
y_line = model.predict(x_line)                 # Calcola la retta di regressione
plt.plot(x_line, y_line, color="red", label='Regr. Lineare')
plt.xlabel('Ore di Studio')
plt.ylabel('Voto Esame')
plt.title('Regressione Lineare: Ore di Studio vs Voto Esame')
plt.legend()
plt.show()
```
