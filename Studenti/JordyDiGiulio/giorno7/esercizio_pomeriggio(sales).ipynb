```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


'''Esercizio 1: Analisi delle Vendite per Categoria
🎯 Obiettivo: Analizzare le vendite totali per categoria di prodotto e visualizzare i risultati.

📋 Istruzioni per la consegna:

Carica il dataset: Usa il file sales_sample.csv per caricare i dati.

Raggruppa i dati per categoria:

Raggruppa i dati per categoria di prodotto (Category).

Calcola la somma totale delle vendite e la somma totale dei ricavi per ciascuna categoria.

Visualizza i risultati:

Crea un grafico a barre che mostri le vendite totali per categoria di prodotto.'''

df = pd.read_csv(r'C:\Users\poppo\Desktop\riverloop project\DataAnalyst-course\Studenti\JordyDiGiulio\giorno7\sales_sample.csv')

categories = df.groupby('Category')[['Sales', 'Revenue']].sum().reset_index().sort_values(by='Sales', ascending=False)
categories

plt.figure(figsize=(8,6))

sns.barplot(
    data=categories,
    x='Category',
    y='Sales',
    hue='Category',
    palette='pastel'
)

plt.title('Vendite totali per categoria')
plt.xlabel('Categorie')
plt.ylabel('Vendite')
plt.tight_layout()
plt.show()
```

```python
'''Esercizio 2: Analisi delle Vendite Mensili
🎯 Obiettivo: Analizzare l'andamento delle vendite e dei ricavi su base mensile.

📋 Istruzioni per la consegna:

Carica il dataset: Usa il file sales_sample.csv per caricare i dati.

Estrai il mese dalla colonna Date:

Estrai l'anno e il mese dalla colonna Date per ottenere i dati su base mensile.

Raggruppa i dati per mese:

Raggruppa i dati per mese (anno e mese).

Calcola la somma delle vendite e la somma dei ricavi per ogni mese.

Visualizza i risultati:

Crea un grafico a linee che mostri l'andamento delle vendite mensili e dei ricavi mensili.'''



df = pd.read_csv('sales_sample.csv')


df['Date'] = pd.to_datetime(df['Date'])


df['anno_mese'] = df['Date'].dt.to_period('M') 


df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')


df_mensile = df.groupby('anno_mese')[['Sales', 'Revenue']].sum().reset_index()


df_mensile['anno_mese'] = df_mensile['anno_mese'].dt.to_timestamp()

# 7. Visualizza i risultati: grafico a linee
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_mensile, x='anno_mese', y='Sales', label='Vendite')
sns.lineplot(data=df_mensile, x='anno_mese', y='Revenue', label='Ricavi')
plt.title('Andamento mensile di Vendite e Ricavi')
plt.xlabel('Mese')
plt.ylabel('Totale')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
'''"Esercizio Voti"Esercizio 3: Top 5 Prodotti per Ricavi
🎯 Obiettivo: Identificare i top 5 prodotti con i maggiori ricavi e visualizzare i risultati.

📋 Istruzioni per la consegna:

Carica il dataset: Usa il file sales_sample.csv per caricare i dati.

Rimuovi valori mancanti nelle colonne Sales e Revenue.

Ordina i dati:

Ordina il dataset in base ai ricavi in ordine decrescente.

Seleziona i top 5 prodotti con i maggiori ricavi.

Visualizza i risultati:

Crea un grafico a barre che mostri i top 5 prodotti con il maggior ricavo.'''

df_pulita = df.dropna()

df_ordinata = df_pulita.sort_values(by='Revenue', ascending=False)

df_plottata = df_ordinata.head(5)

plt.figure(figsize=(8,6))

sns.barplot(

    data=df_plottata,
    x='Product',
    y='Revenue',
    hue='Product',
    palette='pastel'
    
)

plt.title('Top 5 prodotti con maggior ricavo')
plt.xlabel('Prodotti')
plt.ylabel('Ricavi')
plt.legend()
plt.tight_layout()
plt.show()
```
